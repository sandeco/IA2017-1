```python
df.drop(df[df['NOTAS'].str.contains("/",na=False)].index, inplace=True)


df['NOTAS']= df['NOTAS'].astype('float64')

df.loc[df.NOTAS.isnull(),'NOTAS'] = 0


df.NOTAS.unique()
```

```python
def maior_que_dez(x):
    if(x>10):
        return x/10
    else:
        return x

    
    
df.NOTAS = df.loc[:,'NOTAS'].apply(maior_que_dez)  



df.NOTAS.unique()
```
